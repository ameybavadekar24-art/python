### 1. Print numbers from 1 to 10 using a for loop

```python
for num in range(1,11):
    print(num)
```

### 2. Print all even numbers between 1 and 20 using a for loop

```python
for num in range(0,21,2):
    print(num)


for num in range(0,21):      # using if condition print all even number between 1 to 20
    if num % 2==0:
        print(num)
```

### 3. Print each character of the string 'Python' using a for loop

```python
str = "python"
for char in str:
    print(char)
```

### 4. Print the squares of numbers from 1 to 5 using a for loop

```python
for num in range(1,6):
    print(num**2)
```

### 5. Print all elements of a list using a for loop

```python
lst=["Books","Pen","Laptop"]
for element in lst:
    print(element)
```

### 6. Calculate the sum of numbers from 1 to 50 using a for loop

```python
total = 0
for num in range(1,51):
    total+=num
    print(total)
```

### 7. Print numbers from 10 to 1 in reverse using a for loop

```python
for num in range(10,0,-1):
    print(num)
```

### 8. Print the multiplication table of 7 using a for loop

```python
for num in range(11):
    print("7x",num,"=",num*7)
```

### 9. Count how many vowels are in the string 'education' using a for loop

```python
count=0
vowels="aeiou"
for v in "education":
    if v in vowels:
        count += 1
print("total vowels:",count)
```
