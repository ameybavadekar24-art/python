# Basic List Operations 

1. Create a list of 5 numbers and add a new number to the end.


```python
lst=[1,2,3,4,3]
lst.append(24)
lst
```

2. Make a list of fruits and insert a fruit at the second position.


```python
lst = ["mango","apple","gua","orange"]
lst.append("watermelon")
lst
```

3. Create a list of names and remove one name from the list.


```python
lst = ["amey","rahul","akkshhey","prathmesh"]
lst.remove("prathmesh")
lst
```

4. Create a list of 4 items and change the value at index 2.


```python
lst = ["mango","apple","gua","orange"]
lst[2]="banana"
lst
```

5. Make a list of numbers and print only the first three elements.


```python
lst = [1,2,3,4,5]
lst[0:3]
```

6. Create a list of colors and find the index of a specific color.


```python
colors = ["blue","red","pink","purple"]
colors.index("pink")
```

7. Create a list of 10 numbers and print how many times “5” appears.


```python
num = [1,2,3,4,5,6,5,8,9,10]
num.count(5)
```

8. Make a list of 6 items and delete the item at index 3.


```python
item = ["tv","mob","books","pen","laptop","glasses"]
item.remove("books")
item
```

```python
animals = ["lion","tiger","dog"]
animals.append("elephant")
animals.append("buffallow")
animals.append("cow")
animals
```

10. Create a list and remove the last item using pop().


```python
item = ["tv","mob","books","pen","laptop","glasses"]
item.pop()
item
```

11. Create a list of prices and sort them in ascending order.


```python
prices = [10,60,40,20,30,70]
prices.sort()
prices
```

12. Make a list of countries and reverse the order of the list.


```python
countries = ["India","USA","Rassia","Itly"]
countries.reverse()
countries
```

13. Create a list of words and check if “apple” is in the list.

14. Create a list of numbers and print the largest and smallest number.


```python
num = [1,2,3,4,5,6]
print(max(num))
print(min(num))
```

15. Make a list of 5 items and print each item using a loop.


```python
items = ["tv","mob","books","pen","laptop"]
for i in items:
    print(i)
```

```python
items = ["tv","mob","books","pen","laptop"]
fruits = ["mango","apple","gua","orange"]

items.extend(fruits)
items
```

18. Create a list of numbers and print only the even numbers.


```python
num = [1,2,3,4,5,6]
for i in num:
    if i%2==0:
        print(i)
```
