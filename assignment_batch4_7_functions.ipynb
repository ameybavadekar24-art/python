### Write an adapter function that takes a list of operations and executes add, sub, mul, div based on input. 

### Example: [ {"op": "add", "a": 10, "b": 20}, {"op": "mul", "a": 4, "b": 5} ]

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b


def adapter(operations):
    for item in operations:
        op=item["op"]
        a=item["a"]
        b=item["b"]

        if op =="add":
            print(add(a,b))
        elif op == "sub":
            print(sub(a,b))
        elif op == "mul":
            print(mul(a,b))
        elif op == "div":
            print(div(a,b))
        else:
            print("invalid operatios",op)

ops=[{"op":"add","a":10,"b":20},
            {"op":"mul","a":10,"b":20}
        ]


adapter(ops)
```

### Write a function process_students(data) that prints total marks and grade for each student.

```python
data=[ {"name":"A", "marks":[80,90,70]}, {"name":"B", "marks":[60,75,65]} ] 
def process_student(data):
    for student in data:
        name=student["name"]
        marks=student["marks"]
        total_marks=sum(marks)

        print(f"name :{name}")
        print(f"total marks :{total_marks}")

        if total_marks>200:
            print("grade : A")
        else:
            print("grade : B")
        print() 

process_student(data)       
        
        

```

### Create a function apply_discounts(products) that updates price after discount. 
- Example: [ {"name":"Laptop", "price":50000, "discount":10}, {"name":"Phone", "price":20000, "discount":5} ] 
- Expected Output: Laptop final: 45000 Phone final: 19000

```python
def apply_discounts(products):
    for item in products:
        name=item["name"]
        price=item["price"]
        discount=item["discount"]

        discount__price=price*(discount/100)
        final_price=price-discount__price

        print(f"{name} final price :{int(final_price)}")



products = [
    {"name": "Laptop", "price": 50000, "discount": 10},
    {"name": "Phone", "price": 20000, "discount": 5}
]

apply_discounts(products)
```

### Write out_of_stock(items) that returns all products with stock < 20.

```python
def out_of_stock(items):
    for stock in items:
        quantity=stock["quantity"]
        item=stock["item"]
        if quantity<20:
            print(f"{item} : {quantity}")          


stock_items = [
    {"item": "Rice", "quantity": 50},
    {"item": "Wheat", "quantity": 30},
    {"item": "Sugar", "quantity": 20},
    {"item": "Oil", "quantity": 10},
    {"item": "Salt", "quantity": 5}
]

out_of_stock(stock_items)
```

### Write search_product(items, name) â†’ return matching product(s).

```python
def search_product(items,name):
    result=[]

    for item in items:
        if item["item"].lower()==name.lower():

            result.append(item)

            return result

inventory = [
    {"item": "Rice", "price": 50, "quantity": 100},
    {"item": "Wheat", "price": 40, "quantity": 80},
    {"item": "Sugar", "price": 45, "quantity": 60},
    {"item": "Milk", "price": 25, "quantity": 30},
]

print(search_product(inventory, "Rice"))
```
